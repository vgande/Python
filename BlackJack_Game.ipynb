```python
# Import Random to be used to pick a random card from the Deck of Cards
import random
from IPython.display import clear_output

#Global Variables
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,
          'Queen':10,'King':10,'Ace':11}

class Cards():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return f"{self.rank} of {self.suit}"
    
class Deck():
    
    all_cards = []
    
    def __init__(self):
        for suit in suits:
            for rank in ranks:
                createdcard = Cards(suit,rank)
                self.all_cards.append(createdcard)
                
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def remove_one(self):
        return self.all_cards.pop(0)
        
class Player():
    
    def __init__(self,name):
        
        self.name = name
        self.all_cards = []
        
    def add_cards(self,newcard):
        
        self.all_cards.append(newcard)
        
    def __str__(self):
        return f"Number of Cards with Player {self.name} : {len(self.all_cards)}"
    
def print_cards(player_name,flag):
    
    clear_output()
    
    if flag == 'Y':
        
        player_one_count = 0
        print("\nDealer Cards are")
        for num in player_one.all_cards:
            print(player_one.all_cards[player_one_count])
            player_one_count += 1
    else:
        print("Dealer Cards are")
        print("Closed Card")
        print(player_one.all_cards[0])

    print(f"\n{player_name} Cards are")
      
    player_two_count = 0
    
    for num in player_two.all_cards:
        print(player_two.all_cards[player_two_count])
        player_two_count += 1    
        
def value():
    
    player_one_card_value = 0
    player_two_card_value = 0
    player_two_count = 0
    player_one_count = 0
    
    for num in player_one.all_cards:
        player_one_card_value += player_one.all_cards[player_one_count].value
        player_one_count += 1
        
    for num in player_two.all_cards:
        player_two_card_value += player_two.all_cards[player_two_count].value
        player_two_count += 1
        
    return [player_one_card_value,player_two_card_value]
    
def player_two_next_turn(player_two_name):
    
    player_two_on = True
    
    while player_two_on:
        
        player_two_ans = input("Do you want to draw another card: Y or N")
        
        if player_two_ans.upper() not in ['Y','N']:
            print("Enter a valid number")
            
        if player_two_ans.upper() == 'N':
            player_two_on = False
            [player_one_value,player_two_value] = value()
        elif player_two_ans.upper() == 'Y':
            player_two.add_cards(deck_of_cards.remove_one())
            [player_one_value,player_two_value] = value()
            
            if player_two_value > 21:
                print_cards(player_two_name,'Y')
                print(f"\n{player_two_name} is BUSTED. Dealer WINS!!!")
                player_two_on = False
                exit(0)
            else:
                print_cards(player_two_name,'N')
                
    return player_two_value
                
def player_one_next_turn(player_two_name, val):
    
    player_one_value = val
    
    while player_one_value < 18:
        
            player_one.add_cards(deck_of_cards.remove_one())
            [player_one_value,player_two_value] = value()
            
            if player_one_value > 21:
                print_cards(player_two_name,'Y')
                print(f"\nDealer is BUSTED. {player_two_name} WINS!!!")
                player_one_on = False
                exit(0)
            else:
                print_cards(player_two_name,'N')
                
    return player_one_value

#Game Logic
player_one_name = "Dealer"
player_two_name = input("Please enter your name")

player_one = Player(player_one_name)
player_two = Player(player_two_name)

deck_of_cards = Deck()
deck_of_cards.shuffle()

for num in range(2):
    player_one.add_cards(deck_of_cards.remove_one())
    player_two.add_cards(deck_of_cards.remove_one())

#Print the Cards after the initial round
print_cards(player_two_name,'N')

#Collect all the cards for Player
    
player_two_val = player_two_next_turn(player_two_name)

[player_one_value,player_two_value] = value()

if player_one_value >= 18:
    print_cards(player_two_name,'Y')
    new_player_one_value = player_one_value
else:
    new_player_one_value = player_one_next_turn(player_two_name,player_one_value)
    
if player_two_value <=21 and new_player_one_value<=21 and player_two_value > new_player_one_value:
    print_cards(player_two_name,'Y')
    print(f"\n{player_two_name} WINS")
elif player_two_value <=21 and new_player_one_value<=21 and new_player_one_value > player_two_value:
    print_cards(player_two_name,'Y')
    print("\nDealer WINS")
elif player_two_value <=21 and new_player_one_value<=21 and new_player_one_value == player_two_value:
    print_cards(player_two_name,'Y')
    print("\nIT's a TIE")
```
