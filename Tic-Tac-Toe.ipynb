```python
from IPython.display import clear_output

def display_board(board):
    clear_output()  # Remember, this only works in jupyter!
    
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    
def player_letter():
    
    letter_choice = "Z"
    
    while letter_choice not in ['X', 'O']:
        
        letter_choice = input("Please choose a letter in X or O").upper()
        
        if letter_choice == 'X':
            result = ['X','O']
        else:
            result = ['O','X']
            
    return result

def gameon_choice():
    
    choice = False
    
    while not choice:
        
        player_gameon_choice = input("Would you like to play: Y or N?").upper()
        
        if player_gameon_choice not in ['Y','N']:
            print("Incorrect option selected. Please select Y or N")
        else:
            choice = True
            
    return player_gameon_choice

def pick_position(board):
    
    position = False
    
    while not position:
        
        position_number = input("Pick a position between 1-9")
        
        if not position_number.isdecimal():
            print("Enter a valid integer value between 1 and 9")
        elif int(position_number) not in range(1,10):
            print("Invalid position selected: Select a value between 1 and 9")
        elif board[int(position_number)] != ' ':
            print("That position is already filled. Please pick an empty position")
        else:
            position = True
            
    return int(position_number)

def replace_position(board,position_number,position_value):
    
    board[position_number] = position_value
    
    return board

def determine_win_tie(player_letter, board):
    
    win_combination = player_letter*3
    
    horizantal1 = board[1]+board[2]+board[3]
    horizantal2 = board[4]+board[5]+board[6]
    horizantal3 = board[7]+board[8]+board[9]
            
    vertical1 = board[1]+board[4]+board[7]
    vertical2 = board[2]+board[5]+board[8]
    vertical3 = board[3]+board[6]+board[9]
    
    diagonal1 = board[1]+board[5]+board[9]
    diagonal2 = board[3]+board[5]+board[7]
    
    if win_combination in [horizantal1,horizantal2,horizantal3,vertical1,vertical2,vertical3,diagonal1,diagonal2]:
        display_board(board)
        print("----------------")
        print("CONGRATS YOU WIN")
        print("----------------")
        return True
    elif ' ' not in board:
        display_board(board)
        print("----------------")
        print("Well played! It's a Tie")
        print("----------------")
        return True
    else:
        return False
    
def play_game():
    
    game_choice = 'Y'
    
    while game_choice == 'Y':
        
        board = ['a ',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    
        win_draw = False
        game_choice = 'Y'
        count = 1
    
        [player1_letter,player2_letter] = player_letter()
    
        while (not win_draw):
        
            display_board(board)
            pos_num = pick_position(board)
        
            if count%2 == 0:
                playerletter = player2_letter
                count += 1
            else:
                playerletter = player1_letter
                count += 1
        
            current_board = replace_position(board,pos_num,playerletter)
            win_draw = determine_win_tie(playerletter,board)
            
        game_choice = gameon_choice()
            

play_game()
```
